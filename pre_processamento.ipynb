```python
import PyPDF2
import spacy
from nltk.corpus import stopwords
from prettytable import PrettyTable

nlp = spacy.load("pt_core_news_sm")
```

```python
# stopwords
stopwords = stopwords.words('portuguese')
```

```python
def _get_text_from_pdf(pdf_path: str) -> str:
    text = ""
    read_pdf = PyPDF2.PdfReader(pdf_path)
    number_of_pages = len(read_pdf.pages)
    for page_num in range(number_of_pages):
        page = read_pdf.pages[page_num]
        content = page.extract_text()
        doc = nlp(content)
        text += doc.text + "\n"
            
    return text
```

```python
pdf_path = 'documentos/2023/92-2023 - Alteração da Resolução Consuni nº 07-2019 - Prevenção e Segurança Laboratórios da UFT - Consuni-UFT.pdf'
pdf_path_ ='documentos/2023/91-2023 - Regimento Interno da Audin-UFT (Revoga a Resolução Consuni nº 23-2014) - Consuni-UFT.pdf'
texto = _get_text_from_pdf(pdf_path=pdf_path_)

print(texto)
```

```python
doc = nlp(texto)
# print(doc)
text_formatted = spacy.tokens.Doc(doc.vocab, words=[token.text.lower() for token in doc if not token.is_space and not token.is_punct and not token.is_stop])

t = PrettyTable(['Tokens'])
for token in text_formatted:
    t.add_row([token.text])
    
print(t)
```
